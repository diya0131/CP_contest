## 1. SETUP & DATAFRAME EXPLORATORY

```python
# Load libraries & DataFrame
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

path = 'foodprices5nations.csv'
# cols = ["Country ", "Year ", "Month ", "Food_Item ", "Unit of Measurement ", "Price in USD "]
# dtypes = {"Country ": "object", "Year ": "Int32", "Month ": "Int16", "Food Item ": "object", "Unit of Measurement ": "object", "Price in USD ": "float64"}
# df = pd.read_csv(path, usecols=cols, dtype=dtypes)
df = pd.read_csv(path)
df.head()
```

```python
# Remove any white spaces in columns
df.columns = df.columns.str.strip()
# Display column names
df.columns
```

```python
# Columns Removal
df.drop(['Average Price', 'Currency', 'Availability', 'Quality'], inplace=True, axis=1)
df.head()
```

```python
# Datatype Conversion
df=df.astype({"Year": "int32", "Month": "int16"})
df.head()
```

```python
# Info and memory usage check
df.info(memory_usage="deep")
```

```python
# Display the number of entries (number of rows and columns)
rows,cols = df.shape
print(f'Number of rows: {rows}')
print(f'Number of columns: {cols}')
```
